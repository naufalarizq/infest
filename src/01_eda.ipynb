# 01. Exploratory Data Analysis (EDA)
Tahap Exploratory Data Analysis (EDA) sesuai instruksi untuk menemukan *insight* bisnis dan persiapan pemodelan prediksi kesuksesan UMKM.

### Configuration Paths
Konfigurasi direktori di awal cell agar *environment* lebih *compatible*, independen dari OS, dan dapat dijalankan di mana saja dengan struktur proyek yang sama.

```python
import os
from pathlib import Path

# Mengambil base directory secara dinamis (asumsi notebook dijalankan dari dalam folder `src/`)
BASE_DIR = Path.cwd().parent.resolve()
DATA_DIR = BASE_DIR / 'data'
RESULTS_DIR = BASE_DIR / 'results'

RESULTS_DIR.mkdir(parents=True, exist_ok=True)
DATA_PATH = str(DATA_DIR / 'umkm_success.csv')
RESULTS_PATH = str(RESULTS_DIR)

print(f"BASE_DIR    : {BASE_DIR}")
print(f"DATA_PATH   : {DATA_PATH}")
print(f"RESULTS_PATH: {RESULTS_PATH}")
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

# Load dataset menggunakan path dinamis
df = pd.read_csv(DATA_PATH)
```

### 1. Tampilkan 5 Data Teratas, Info Data, dan Deskripsi Data

```python
display(df.head())
print("\n--- INFO DATA ---")
df.info()
print("\n--- DESKRIPSI DATA ---")
display(df.describe())
```

### 2. Pengecekan Data Duplikat, Missing Value, dan Unique Value

```python
dup_count = df.duplicated().sum()
dup_pct = (dup_count / len(df)) * 100
print(f"Jumlah data duplikat: {dup_count} ({dup_pct:.2f}%)\n")

data_quality = pd.DataFrame({
    'Missing Value': df.isnull().sum(),
    'Percentage Missing (%)': (df.isnull().sum() / len(df)) * 100,
    'Unique Value': df.nunique()
})

unique_samples = []
for col in df.columns:
    if df[col].nunique() <= 10:
        unique_samples.append(", ".join(map(str, df[col].unique())))
    else:
        unique_samples.append("-")

data_quality['Unique Data (<= 10)'] = unique_samples
display(data_quality)
data_quality.to_csv(f'{RESULTS_PATH}/data_quality_summary.csv')
```

### 3. Distribusi Data Target (Success)

```python
fig_target = px.pie(df, names='Success', title='Distribusi Target: Success (0 = Tidak Berhasil, 1 = Berhasil)', hole=0.3)
fig_target.show()
fig_target.write_html(f'{RESULTS_PATH}/distribusi_success.html')
```

### 4. Distribusi Data Numerik & Analisis Skewness

```python
numeric_cols = ['Age', 'Industry_Experience', 'Education', 'Marketing_Effort', 'Professional_Advice']

for col in numeric_cols:
    skewness = df[col].skew()
    fig = px.histogram(df, x=col, title=f'Distribusi {col} (Skewness: {skewness:.2f})', marginal='box', text_auto=True)
    fig.show()
    fig.write_html(f'{RESULTS_PATH}/distribusi_numerik_{col}.html')
```

### 5. Distribusi Data Kategorik/Ordinal

```python
categorical_cols = ['Education', 'Initial_Capital', 'Financial_Record_Keeping', 
                    'Internet_Usage', 'Business_Plan', 'Marketing_Effort', 
                    'Partnership', 'Parent_Business_Experience', 'Owner_Gender', 
                    'Professional_Advice']

for col in categorical_cols:
    val_counts = df[col].value_counts().reset_index()
    val_counts.columns = [col, 'Count']
    val_counts[col] = val_counts[col].astype(str)
    fig = px.bar(val_counts, x=col, y='Count', title=f'Distribusi Kategorik: {col}', text='Count')
    fig.show()
    fig.write_html(f'{RESULTS_PATH}/distribusi_kategorik_{col}.html')
```

### 6. Heatmap Korelasi
- Korelasi Pearson untuk fitur Numerik
- Korelasi Spearman untuk fitur Kategorik/Ordinal

```python
# Hitung korelasi untuk seluruh kolom (termasuk variabel 'Success')
corr_all = df.corr()

# Tampilkan heatmap korelasi keseluruhan
fig_corr = px.imshow(
    corr_all, 
    text_auto=".2f", 
    title='Heatmap Korelasi Keseluruhan Fitur dan Target', 
    aspect='auto',
    color_continuous_scale='RdBu_r' # Skala warna merah (negatif) ke biru (positif)
)

# Sesuaikan ukuran agar lebih nyaman dibaca karena variabelnya banyak
fig_corr.update_layout(height=800, width=800)

fig_corr.show()
fig_corr.write_html(f'{RESULTS_PATH}/korelasi_keseluruhan.html')
```

### 7. Deteksi Outlier pada Kolom Numerik menggunakan IQR

```python
outlier_summary = {}
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_summary[col] = len(outliers)
    
    # Visualisasi boxplot dengan Plotly
    fig = px.box(df, y=col, title=f'Boxplot {col} (Jumlah Outlier: {len(outliers)})')
    fig.show()
    fig.write_html(f'{RESULTS_PATH}/boxplot_outlier_{col}.html')

print("--- RINGKASAN JUMLAH OUTLIER ---")
for col, count in outlier_summary.items():
    print(f"{col}: {count}")
```

```python
print("--- 1. CEK KONSISTENSI LOGIS ---")
# Asumsi usia mulai kerja logis paling cepat adalah 15 tahun
inconsistent_exp = df[df['Age'] - df['Industry_Experience'] < 15]
print(f"Jumlah data tidak logis (Usia saat mulai industri < 15 tahun): {len(inconsistent_exp)} baris")
if len(inconsistent_exp) > 0:
    display(inconsistent_exp[['Age', 'Education', 'Industry_Experience']].head())

inconsistent_edu = df[(df['Education'] >= 4) & (df['Age'] < 14)]
print(f"\nJumlah data tidak logis (Pendidikan tinggi tapi umur < 14 tahun): {len(inconsistent_edu)} baris")
if len(inconsistent_edu) > 0:
    display(inconsistent_edu[['Age', 'Education']].head())


print("\n--- 2. CEK RENTANG NILAI (DOMAIN) ---")
# Cek Skala Likert
invalid_marketing = df[~df['Marketing_Effort'].between(1, 7)]
invalid_advice = df[~df['Professional_Advice'].between(1, 7)]
print(f"Marketing Effort di luar rentang 1-7: {len(invalid_marketing)} baris")
print(f"Professional Advice di luar rentang 1-7: {len(invalid_advice)} baris")

# Cek Variabel Biner
binary_cols = ['Initial_Capital', 'Financial_Record_Keeping', 'Internet_Usage', 
               'Business_Plan', 'Partnership', 'Parent_Business_Experience', 
               'Owner_Gender', 'Success']
for col in binary_cols:
    invalid_bin = df[~df[col].isin([0, 1])]
    if len(invalid_bin) > 0:
        print(f"WARNING: Kolom biner '{col}' memiliki nilai selain 0 dan 1!")
```
